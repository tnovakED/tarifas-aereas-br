```python
from pyspark.sql.functions import current_timestamp, expr
```

```python
df_cias_aereas = spark.table("`ct-esteira-dados-aviacao`.raw.fat_voos_raw_dep_arr")
```

```python
# seleciona as empresas aereas em uma tabela dimensao.

df_cias_aereas = df_cias_aereas.select("sigla_cia_aerea", "empresa_aerea").distinct().orderBy("sigla_cia_aerea")

# adiciona um coluna timestamp para controle de carga.

df_cias_aereas = df_cias_aereas.withColumn("load_data", expr("current_timestamp() - interval 3 hours"))
display(df_cias_aereas)
```

```python
df_cias_aereas \
    .write \
    .format('delta') \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .partitionBy("sigla_cia_aerea") \
    .saveAsTable("`ct-esteira-dados-aviacao`.raw.dim_cias_aereas")
```
