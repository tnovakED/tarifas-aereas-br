```python
from pyspark.sql.functions import current_timestamp, expr
```

### Trocar os nomes das colunas

```python
# Leitura da tabela delta
dim_aero_siglas_trusted = spark.read.table("`ct-esteira-dados-aviacao`.raw.dim_aero_siglas")

# mudança de nome da coluna
dim_aero_siglas_trusted = dim_aero_siglas_trusted \
.withColumnRenamed("codigo_icao", "s_aero_icao")

# mudança de nome da coluna
dim_aero_siglas_trusted = dim_aero_siglas_trusted \
.withColumnRenamed("codigo_iata", "s_aero_iata")

# adicionar uma coluna de controle de carga
dim_aero_siglas_trusted = dim_aero_siglas_trusted \
.withColumn("load_data", expr("current_timestamp() - interval 3 hours"))
```

```python
dim_aero_siglas_trusted.write \
    .mode("overwrite") \
    .format("delta") \
    .option("overwriteSchema", "true") \
    .partitionBy("s_aero_iata") \
    .saveAsTable("`ct-esteira-dados-aviacao`.trusted.dim_aero_siglas_truested")
```
